### Loading and Filtering the Dataset

This section:
1. **Loads the Dataset**: Ensures that all the data is properly imported into a DataFrame for analysis.
2. **Date Conversion**: Converts the `Date` column into a datetime format to facilitate time-based filtering.
3. **Filter Relevant Data**: Focuses on data from 2005 onward, reducing noise and ensuring the model only uses recent and relevant data. This enhances the model's efficiency by providing a clean dataset without unnecessary historical information.

By filtering and preprocessing the dataset, the model can better capture patterns in property prices over time, improving accuracy and reducing overfitting caused by irrelevant data.


````python
# Ensure df_cleaned is an independent copy to avoid SettingWithCopyWarning
df_cleaned = df[df['Date'].dt.year >= 2005].copy()

# Extract year, quarter, and month from 'Date' column
df_cleaned['Year'] = df_cleaned['Date'].dt.year
df_cleaned['Quarter'] = df_cleaned['Date'].dt.quarter
df_cleaned['Month'] = df_cleaned['Date'].dt.month

# Calculate rolling averages (12-month rolling average as an example)
df_cleaned['Detached_Rolling_Avg'] = df_cleaned['Detached_Average_Price'].rolling(window=12).mean()
df_cleaned['Semi_Detached_Rolling_Avg'] = df_cleaned['Semi_Detached_Average_Price'].rolling(window=12).mean()
df_cleaned['Terraced_Rolling_Avg'] = df_cleaned['Terraced_Average_Price'].rolling(window=12).mean()
df_cleaned['Flat_Rolling_Avg'] = df_cleaned['Flat_Average_Price'].rolling(window=12).mean()

# Group by 'Region_Name' and 'Year' and calculate the mean
region_avg_prices = df_cleaned.groupby(['Region_Name', 'Year'])[
    ['Detached_Average_Price', 'Semi_Detached_Average_Price',
     'Terraced_Average_Price', 'Flat_Average_Price']
].mean().reset_index()

# Display the new columns to check the changes
print(df_cleaned[['Year', 'Quarter', 'Month', 'Detached_Rolling_Avg',
                  'Semi_Detached_Rolling_Avg', 'Terraced_Rolling_Avg', 'Flat_Rolling_Avg']].head())
````

### Time-Based Feature Extraction and Rolling Averages

This section:
1. **Time-Based Features**:
   - Extracts the `Year`, `Quarter`, and `Month` from the `Date` column to add time-based granularity, which improves the model's understanding of seasonal and yearly patterns.

2. **Rolling Averages**:
   - Calculates 12-month rolling averages for property prices. Rolling averages help the model focus on smoothed trends rather than short-term fluctuations, improving prediction stability and reducing overfitting.

3. **Grouping Data**:
   - Aggregates average property prices by `Region_Name` and `Year`. This step creates region-specific summaries, enabling the model to learn regional differences in property prices, which enhances prediction accuracy.

These steps ensure the model captures time-based trends, smoothens volatility in price data, and identifies regional variations in property prices.


````python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Filter data for years 2005-2007 for training and 2008 for testing
train_data = df_cleaned[df_cleaned['Year'] < 2008]
test_data = df_cleaned[df_cleaned['Year'] == 2008]

# Define features and target variable
features = ['Detached_Average_Price', 'Semi_Detached_Average_Price',
            'Terraced_Average_Price', 'Flat_Average_Price', 'Year', 'Quarter', 'Month',
            'Detached_Rolling_Avg', 'Semi_Detached_Rolling_Avg', 'Terraced_Rolling_Avg', 'Flat_Rolling_Avg']
target = 'Detached_Average_Price'  # Target variable is the price for detached houses

# Drop rows with missing values in features or target
train_data = train_data.dropna(subset=features + [target])
test_data = test_data.dropna(subset=features + [target])

# Separate features and target for training and testing
X_train = train_data[features]
y_train = train_data[target]
X_test = test_data[features]
y_test = test_data[target]

# Train the RandomForestRegressor model
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)

# Predict on the test data (2008)
y_pred = model.predict(X_test)

# Evaluate the model using metrics: RMSE, MAE, and R²
rmse = mean_squared_error(y_test, y_pred, squared=False)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

rmse, mae, r2
````
